## Hypothesis Testing

**Test if there's a significant difference in bike rentals between weekdays and weekends**

```python
#1st I have to create a new column for analysis

# Group by 'day' and calculate the sum of counts
daily_counts = train_data.groupby('day')['count'].sum().reset_index()

# Create a new column 'day_type' indicating 'weekday' or 'weekend'
daily_counts['day_type'] = daily_counts['day'].apply(lambda x: 'weekday' if x in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'] else 'weekend')

daily_counts
```

```python
#performing z-test as it is two-sample, two-sided 
from statsmodels.stats.weightstats import ztest

# Creating variables that have the count per weekday and weekend
weekday= daily_counts[daily_counts.day_type == 'weekday']['count']
weekend = daily_counts[daily_counts.day_type == 'weekend']['count']

# Perform the z-test
statistic, p_value = ztest(weekday, weekend)

# Display the results
print(f"Z-Statistic: {statistic}")
print(f"P-Value: {p_value}")


#Result indicates that the weekday mean is smaller than weekend mean
```

```python
#Null hypothesis: Weekday mean is less than or equal to weekend mean
#Smaller to see if weekend bike rentals are greater than weekday

ztest(weekday, weekend, alternative = 'smaller')

##This has a null hypothesis that the weekday mean is larger than or equal to weekend mean. 
#This returns a p-value of 0.45, more than our of 0.05 and a z-stat of -0.1
#So, we accept the null hypothesis and state that weekday mean is smaller than weekend mean with statistical significance. 
```

**Test if the average temperature affects bike rentals**

```python
#Test if the average temperature affects bike rentals - using Pearson correlation test

from scipy.stats import pearsonr

# Creating variables for temepratures and bike rentals
temperature = train_data['temp']
bike_rentals = train_data['count']

# Calculate the Pearson correlation coefficient and p-value
correlation_coefficient, p_value = pearsonr(temperature, bike_rentals)

# Display the results
print(f"Pearson Correlation Coefficient: {correlation_coefficient}")
print(f"P-Value: {p_value}")

# Interpret the results
if p_value < 0.05:
    print("There is a significant correlation between temperature and bike rentals.")
else:
    print("There is no significant correlation between temperature and bike rentals.")
```

```python
from scipy.stats import pearsonr
from pandas.api.types import is_numeric_dtype

# Iterate through each column in the DataFrame
for c in train_data.columns:
    # Check if the column contains numeric data
    if is_numeric_dtype(train_data[c]):
        # Calculate the Pearson correlation coefficient and p-value between the current column and 'count'
        correlation, pvalue = pearsonr(train_data[c], train_data['count'])
        
        # Print the column name, correlation coefficient, and whether it's statistically significant
        print(f'{c : <40}: {correlation: .4f}, significant: {pvalue <= 0.05}')
```

```python
# Create a copy of the DataFrame to work with to create categories
train_data_copy = train_data.copy()

# Define a list of columns to be treated as interval variables
interval_columns = ['temp', 'humidity', 'windspeed', 'count']

# Loop through each column in the DataFrame
for c in train_data_copy.columns:
    # Check if the column is not in the list of interval columns
    if c not in interval_columns:
        # Convert the column to a categorical variable
        train_data_copy[c] = train_data_copy[c].astype('category')
```

```python
train_data_copy.dtypes
```

```python
import phik
from scipy.stats import pearsonr

# Instantiate lists to store phi-k correlations and significances
phik_correlations = []     # To store calculated phi-k correlations
phik_significances = []    # To store calculated significances

# Get the column names of the DataFrame
columns = train_data_copy.columns

# Extract the target variable ('count')
y = train_data_copy['count']

# Loop through each column in the DataFrame for correlation analysis
for c in columns:
    # Extract the current column as the predictor variable 'x'
    x = train_data_copy[c]
    
    # Check if the current column is an interval variable
    if c in interval_columns:
        # Calculate the Pearson correlation for interval variables
        correlation, p_value = pearsonr(x, y)
        phik_correlations.append(correlation)
        phik_significances.append(p_value)
    else:
        # Calculate the phi-k correlation for categorical variables and append to the correlations list
        phik_correlations.append(phik.phik_from_array(x, y))

        # Calculate the significance of the correlation and append to the significances list
        # Currently, set drop over/underflow to false to avoid an error (expected to be fixed in an update)
        phik_significances.append(phik.significance.significance_from_array(x, y, drop_overflow=False, drop_underflow=False)[0])
```

```python
# Create a DataFrame to store phi-k correlations and p-values
phik_df = pd.DataFrame({'phik': phik_correlations, 'p-value': phik_significances}, index=columns)

# Sort the DataFrame based on phi-k correlation values in descending order
phik_df.sort_values(by='phik', ascending=False, inplace=True)
```

```python
# Generate bar plots for phi-k correlations and p-values, excluding the first row (index 0)
# This selects all rows in the phik_df DataFrame starting from the second row (index 1).
phik_df.iloc[1:].plot.bar(subplots=True, figsize=(5.5, 5.5))

# Adjust the layout to ensure the subplots fit within the figure
plt.tight_layout()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd

# Set x and y
X = final_data.drop('count', axis=1)
y = final_data['count']

# Set threshold based on analysis above
threshold = 145  

# Convert count to categorical/binary value
y_categorical_as_binary = (y > threshold).astype(int)

# Split the data into training and testing sets
train_features, test_features, train_targets, test_targets = train_test_split(
    X,
    y_categorical_as_binary,
    test_size=0.2,      # 0.2 for an 80-20 split 
    random_state=42     # Set a random seed for reproducibility
)

# Logistic regression model
lr_sklearn = LogisticRegression()
lr_sklearn.fit(train_features, train_targets)

# Make predictions on the test set
y_pred_prob = lr_sklearn.predict_proba(test_features)[:, 1]

# Apply the threshold to convert probabilities to binary predictions
y_pred = (y_pred_prob > 0.5).astype(int)

# Evaluate the model
accuracy = accuracy_score(test_targets, y_pred)
classification_rep = classification_report(test_targets, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", classification_rep)

```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Assign variable to get all numeric values from final data
cluster_data = final_data[['temp', 'humidity', 'windspeed', 'count']]

# Perform feature scaling using StandardScaler
scaler = StandardScaler()
cluster_data_scaled = scaler.fit_transform(cluster_data)

# Determine the optimal number of clusters using the Elbow method
# Found here: https://www.geeksforgeeks.org/k-means-clustering-introduction/

sse = []  # SUM OF SQUARED ERROR

for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=2)
    km.fit(cluster_data_scaled)  
    sse.append(km.inertia_)

# Plot the Elbow method to find the optimal number of clusters
sns.set_style("whitegrid")
g = sns.lineplot(x=range(1, 11), y=sse)
 
g.set(xlabel="Number of cluster (k)", 
      ylabel="Sum Squared Error", 
      title='Elbow Method')

```

```python
# Based on graph above, we can see that elbow-like situation happens between k=2 and k=3. 
# I consider k=3 as optimal cluster
optimal_clusters = 3

# Perform K-means clustering with the chosen number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=2)

#predict cluster group and store as new column in final data column ['numeric_cluster']
final_data['numeric_cluster'] = kmeans.fit_predict(cluster_data_scaled)

# Visualize the clusters (only considering temperature and bike rentals)
plt.scatter(final_data['temp'], final_data['count'], c=final_data['numeric_cluster'], cmap = 'cividis') #x = temp, y=count, color=numeric cluster with color map cividis
plt.xlabel('Temperature')
plt.ylabel('Bike Rentals')
plt.title('K-means Clustering of Days')
plt.show()
```
